## Calculate Monthly Sales Totals and Display Them in a Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("C:/Users/Aniket/Downloads/codeSamurai/retail_sales_dataset.csv")

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a 'Month' column for grouping
df['Month'] = df['Date'].dt.to_period('M')

# Calculate total sales per month
monthly_sales = df.groupby('Month')['Total Amount'].sum()

# Plotting the monthly sales in a line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales, marker='o')
plt.title('Monthly Sales Totals')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##  Identify the Best and Worst Months for Sales

```python
# Best month (maximum sales)
best_month = monthly_sales.idxmax()
best_sales = monthly_sales.max()

# Worst month (minimum sales)
worst_month = monthly_sales.idxmin()
worst_sales = monthly_sales.min()

print(f"Best Month: {best_month} with sales {best_sales}")
print(f"Worst Month: {worst_month} with sales {worst_sales}")
```

##  Analyze Sales by Product Category Using a Bar Chart

```python
# Analyze sales by Product Category
sales_by_category = df.groupby('Product Category')['Total Amount'].sum()

# Plotting the sales by product category using a bar chart
plt.figure(figsize=(10, 6))
sales_by_category.plot(kind='bar', color='skyblue')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```
